```python
# Filling the BLank data using interpolation method 
year_columns = data.columns[3:]  
data[year_columns] = data[year_columns].apply(pd.to_numeric, errors='coerce')
data[year_columns] = data[year_columns].interpolate(method='linear', axis=1, limit_direction='both')
```
