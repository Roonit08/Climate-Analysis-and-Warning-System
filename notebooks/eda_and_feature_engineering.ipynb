```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
data = pd.read_csv("../data/cleaned_data.csv")
```

```python
data
```

```python
# Creating a list of year columns from 1970 to 2021
year_cols = [str(year) for year in range(1970, 2022)]
```

```python
# Country with Highest rise of Temperature from baseline
max_val = data[year_cols].max().max()
max_year = data[year_cols].max().idxmax()   
max_country = data.loc[data[max_year].idxmax(), "Country Name"]
print("Highest rise value:", max_val)
print("Year:", max_year)
print("Country:", max_country)
```

```python
# Counrty with lowest drop of temperature form baseline
min_val = data[year_cols].min().min()
min_year = data[year_cols].min().idxmin()
min_country = data.loc[data[min_year].idxmin(), "Country Name"]
print("lowest drop value:", min_val)
print("Year:", min_year)
print("Country:", min_country)
```

```python
# Average change in temperature of every year
yearly_mean = data[year_cols].mean()
print(yearly_mean)
```

```python
# Average change in temperature of entire data 
global_avg_all_years = yearly_mean.mean()
print("Global Average change in temperature:",global_avg_all_years)
# Actual Global Change in temperature at present as compared to baseline temperature 
```

```python
# Every countries Average change in temperature over the period 
country_mean = data.set_index("Country Name")[year_cols].mean(axis=1)
print(country_mean)
```

```python
# Country with highest average rise in temperature over the year
highest_country = country_mean.idxmax()
highest_value = country_mean.max()
print(highest_country,highest_value)
```

```python
# Country with lowest average drop in temperature over the year
lowest_country = country_mean.idxmin()
lowest_value = country_mean.min()
print(lowest_country,lowest_value)
```

```python
## Number of countries with the temperature rise above 2 degree celcius 
year_cols = [col for col in data.columns if col.isdigit()]

# Rise relative to 1970 for each year
temp_change = data[year_cols].subtract(data["1970"], axis=0)

# The maximum rise for each country
data["max_rise"] = temp_change.max(axis=1)

# Countries with >2°C rise
countries_over_2 = data[data["max_rise"] > 2]["Country Name"]
print("Number of countries with >2°C rise:", len(countries_over_2))
print("\nCountries:\n", countries_over_2.tolist())
```

```python
# Comparing average temperature change in first half (1970-1995) vs second half (1996-2021) of the period
first_half = [str(y) for y in range(1970, 1996)]
second_half = [str(y) for y in range(1996, 2022)]
first_half_avg = data[first_half].mean().mean()
second_half_avg = data[second_half].mean().mean()
print("Avg change 1970-1995:", first_half_avg)
print("Avg change 1996-2021:", second_half_avg)
```

```python
# Countries with the most volatile (unstable) year-to-year temperature change
country_std = data.set_index("Country Name")[year_cols].std(axis=1)
most_volatile = country_std.sort_values(ascending=False).head(10)
print(most_volatile)
```

```python
# Linegraph of global yearly average temperature change 
plt.figure(figsize=(14,6))
plt.plot(yearly_mean.index, yearly_mean.values, marker="o")
plt.title("Global Yearly Mean Temperature Change (1970–2021)")
plt.xlabel("Year")
plt.ylabel("Temperature Change (°C)")
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

```python
# Bargraph of global yearly average temperature change 
plt.figure(figsize=(14,6))
plt.bar(yearly_mean.index, yearly_mean.values, color="skyblue")
plt.title("Global Yearly Mean Temperature Change (1970–2021)")
plt.xlabel("Year")
plt.ylabel("Temperature Change (°C)")
plt.xticks(rotation=90)
plt.show()
```

```python
# Scattered graph of country wise average temperature change
plt.figure(figsize=(14,6))
plt.scatter(country_mean.index, country_mean.values, alpha=0.7)
plt.title("Country-wise Mean Temperature Change (1970–2021)")
plt.xlabel("Country")
plt.ylabel("Mean Temperature Change (°C)")
plt.xticks([], [])
plt.show()
```

```python
# Line graph of temperature change in all countries over the years 
plt.figure(figsize=(14,8))
for _, row in data.iterrows():
    plt.plot(year_cols, row[year_cols].values, alpha=0.3)
plt.title("Temperature Change Trends of All Countries (1970–2021)")
plt.xlabel("Year")
plt.ylabel("Temperature Change (°C)")
plt.xticks(rotation=90)
plt.show()
```

```python

# Heatmap of countries according to average temperature change in the period 
sorted_countries = country_mean.sort_values(ascending=False)
plt.figure(figsize=(10,38))
plt.imshow(sorted_countries.values.reshape(-1,1), cmap="coolwarm", aspect="auto")
plt.colorbar(label="Mean Temperature Change (°C)")
plt.yticks(range(len(sorted_countries)), sorted_countries.index, fontsize=6)
plt.xticks([0], ["Countries"])
plt.title("Heatmap of Countries by Mean Temperature Change (1970–2021)")
plt.show()
```

```python
# Barchart of top 10 most affected and bottom 10 least affected countries 
top10 = country_mean.sort_values(ascending=False).head(10)
bottom10 = country_mean.sort_values().head(10)
plt.figure(figsize=(14,6))
plt.bar(top10.index, top10.values, color="red", label="Most Affected")
plt.bar(bottom10.index, bottom10.values, color="blue", label="Least Affected")
plt.title("Top 10 and Bottom 10 Countries by Mean Temperature Change (1970–2021)")
plt.xlabel("Country")
plt.ylabel("Mean Temperature Change (°C)")
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

```python
# Boxplot of yearly distribution of temperature change across all the countries
plt.figure(figsize=(14, 6))
plt.boxplot([data[year].dropna().values for year in year_cols])
plt.ylabel("Temperature Change (°C)")
plt.title("Yearly Distribution of Temperature Changes Across Countries")
plt.xticks(range(1, len(year_cols) + 1, 5), year_cols[::5], rotation=90) 
plt.show()
```

```python
# Piechart amoung top 10 most affected countries with temperature change 
top10 = country_mean.sort_values(ascending=False).head(10)
plt.pie(top10, labels=top10.index, autopct='%1.1f%%')
plt.title("Top 10 Most Affected Countries")
plt.show()
```

```python
# Reshaping data from wide format to long format for feature engineering
long_df = data.melt(
    id_vars=["Country Name"],
    value_vars=year_cols,
    var_name="year",
    value_name="temp_anomaly"
)
long_df["year"] = long_df["year"].astype(int)
long_df = long_df.sort_values(["Country Name", "year"])
```

```python
# Quick look at the dataset after adding new columns
long_df["lag_1"] = long_df.groupby("Country Name")["temp_anomaly"].shift(1)
long_df["rolling_mean_5"] = long_df.groupby("Country Name")["temp_anomaly"].transform(lambda x: x.rolling(5, min_periods=1).mean())
long_df["decade"] = (long_df["year"] // 10) * 10
long_df["country_encoded"] = long_df["Country Name"].astype("category").cat.codes
```

```python
# Dropping rows with no previous year to reference (first year per country)
print(long_df["lag_1"].isnull().sum())
long_df = long_df.dropna(subset=["lag_1"])
```
